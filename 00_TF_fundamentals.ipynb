### Creating random tensors.

Random tensors are tensors of some arbitary size which contains random numbers.


Rarely in practice we would have to decide b/w whether to choose tf.constant or tf.Variable as TensorFlow does this for us. If in doubt,use tf.constant and change it later if required.


```python
random_1 = tf.random.Generator.from_seed(42)
random_1=random_1.normal(shape=(3,2))

random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))
random_1,random_2,random_1 == random_2
```

```python
samples = tf.random.normal(shape=(3,2))
samples
```

```python
# Gamma function (has 2 parameters,alpha and beta)

samples = tf.random.gamma([15],alpha=[10],beta=[0.5])
print(samples)
```

```python
alpha = tf.constant([2.5])
print(alpha)
beta = tf.random.normal(shape=(3,2))
print(beta)

samples = tf.random.gamma([10],alpha=alpha,beta=beta)
# 10 is the number of samples that is going to be generated for corresponding alpha and beta

samples
```

### Shuffle the order of the elements of the tensor

```python
# Shuffle a tensor (Valuable when we want to shuffle our data so the inherent order doesnt affect learning)

not_shuffled = tf.constant([[10,7],
                            [3,4],
                            [2,5]])

# Shuffle the above tensor
tf.random.shuffle(not_shuffled) # Every time you run this you will get a different output


```

```python
not_shuffled
```

```python
tf.random.set_seed(42) # Global level random Seed
tf.random.shuffle(not_shuffled) # Operational level random seed

#In order to get the same tensor all the time when we are running this cell
#it is necessary that we declare both the Global and Operational level seeds.

```

### Indexing in tensors

Tensors can be indexed just like Python lists.

```python
# Get the first 2 elements of each dimension
rank_4_tensor[:2,:2,:2,:2]
```

```python
# Get the first element from each dimension from each index except for the final one
rank_4_tensor[:1,:1,:1,:]
```

```python
# Create a rank 2 tensor
tf.random.set_seed(24)
rank_2_tensor  = tf.random.normal([2,2])
rank_2_tensor
```

```python
#Get the last item of each of row of our rank 2 tensor
rank_2_tensor[:,-1]  # Means consider all the rows but only get the values in the last column.
```

```python
# Add in extra dimension to our rank 2 tensor
rank_3_tensor = rank_2_tensor[...,tf.newaxis] # rank_3_tensor = rank_2_tensor[:,:,tf.newaxis]
rank_3_tensor
```

```python
#Alteranate to tf.newaxis
tf.expand_dims(rank_2_tensor,axis=-1) # expands the dimension on the last axis
```

```python
tf.expand_dims(rank_2_tensor,axis=0) # expands the dimension on the 0th axis
```

This is the most common tensor ops.

```python
t3 =tf.constant([[2,2,3],
            [4,5,6]])

t4 = tf.constant([[1,2],
                  [9,10],
                  [7,5]])

tf.matmul(t3,t4).numpy()
```

### Aggregating Tensors

Aggrgating tensors = considering them from multiple values down to a smaller amount of values.

```python
# Get the abosulte values
D = tf.constant([-7,-10])
D
```

```python
tf.abs(D)
```

Let's go through the following forms of Aggregation:
* Get the maximum
* Get the minimum
* Get the mean of the tensor
* Get the sum of the tensor

```python
f = tf.constant([[5,6,7],[8,20,30]])

g = tf.constant([[7,10,11],
                 [10,11,12]])
tf.maximum(f,g)
```

```python
tf.reduce_mean(f).numpy()
```

```python
K = tf.constant(np.random.randint(0,100,size = 50))
K
```

```python
tf.reduce_min(K) # The minimum element in the array
```

```python
tf.reduce_max(K)
```

```python
M = tf.reduce_mean(K)
M
```

```python
tf.reduce_sum(K).numpy()
```

```python
import tensorflow_probability as tfp
tfp.stats.variance(K).numpy()
```

### One Hot Encoding Tensors

```python
# One hot encoding is a technique for numerical encoding,and we need to
# follow this step before we input our data in the Neural Network.

# Create a list of indices

some_list  = [0,1,2,3] # Red,Green,Blue and Purple
tf.one_hot(some_list , depth = 4)
```

```python
# Specify custom values for one hot encoding
tf.one_hot(some_list,depth = 4,on_value = "3",off_value = "2")
```
